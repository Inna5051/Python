```python
#11 Pandas
#Створіть функцію, яка завантажує дані з таблиці Products, множить кожне значення поля UnitPrice на 10, сумує ці значення і повертає середнє значення.
import pandas as pd
from sqlalchemy import create_engine

def task_13():

    # Налаштування підключення
    server = '10.206.0.254\BIGDATA_SQL2019'
    database = 'Northwind'

    # Створення рядка підключення з довіреним з'єднанням
    connection_string = f'mssql+pyodbc://{server}/{database}?trusted_connection=yes&driver=ODBC+Driver+17+for+SQL+Server'

    # Створення об'єкта бази даних
    engine = create_engine(connection_string, echo=False)

    result_df = pd.read_sql_query('SELECT "UnitPrice" FROM Northwind.dbo.Products', engine)
    result_df['UnitPrice'] *= 10
    result_mean = sum(result_df['UnitPrice'])/len(result_df['UnitPrice'])

    #print(result_df)

    return float(result_mean)

print (task_13())
```
