```python
#3 Створіть функцію, яка використовує бібліотеку pandas для створення DataFrame з CSV файлу,
#  виконує сортування даних за одним з стовпців і зберігає результат у новий CSV файл.
import pandas as pd
import csv

def task_3_createCSV():
    filename = 'sales_data1.csv'

    sales_data = {
    'місяць': 'серпень',
    'рік': 2024,
    'продажі': [
        {'дата': '2024-08-01', 'продукт': 'Продукт A', 'кількість': 100, 'вартість': 5000},
        {'дата': '2024-08-02', 'продукт': 'Продукт B', 'кількість': 150, 'вартість': 7500},
        {'дата': '2024-08-03', 'продукт': 'Продукт C', 'кількість': 200, 'вартість': 10000},
        {'дата': '2024-08-04', 'продукт': 'Продукт D', 'кількість': 120, 'вартість': 6000},
        {'дата': '2024-08-05', 'продукт': 'Продукт E', 'кількість': 180, 'вартість': 9000},
        {'дата': '2024-08-06', 'продукт': 'Продукт F', 'кількість': 160, 'вартість': 8000},
        {'дата': '2024-08-07', 'продукт': 'Продукт G', 'кількість': 130, 'вартість': 6500},
        {'дата': '2024-08-08', 'продукт': 'Продукт H', 'кількість': 140, 'вартість': 7000},
        {'дата': '2024-08-09', 'продукт': 'Продукт I', 'кількість': 110, 'вартість': 5500},
        {'дата': '2024-08-10', 'продукт': 'Продукт J', 'кількість': 190, 'вартість': 9500}
    ],}

    # Відкриваємо файл для запису
    with open(filename, 'w', newline='', encoding='utf-8') as csvfile:
        writer = csv.writer(csvfile)
        # Записуємо метадані
        writer.writerow(['місяць', sales_data['місяць']])
        writer.writerow(['рік', sales_data['рік']])
        writer.writerow([])

        # Записуємо заголовки таблиці
        writer.writerow(['дата', 'продукт', 'кількість', 'вартість'])

        # Записуємо дані продажів
        for sale in sales_data['продажі']:
            writer.writerow([sale['дата'], sale['продукт'], sale['кількість'], sale['вартість']])

    result_dict = dict()
    subdict_keys = list()
    is_sales = False
    
    with open('sales_data1.csv', newline='', encoding='utf-8') as csvfile:
        reader = csv.reader(csvfile)

        for row in reader:
            if row == []:
                subdict_keys = next(reader)
                break
            result_dict[row[0]] = row[1]
        
        for row in reader:
            if is_sales:
                if row != subdict_keys:
                    continue
                if row == subdict_keys:
                    is_sales = True
                    continue
            temp_dict =[{
                subdict_keys[0]: row[0],
                subdict_keys[1]: row[1],
                subdict_keys[2]: row[2],
                subdict_keys[3]: row[3]
            },]

            if 'продажі' not in result_dict:
                result_dict['продажі'] = []

            result_dict['продажі'].append(temp_dict)

    df = pd.DataFrame(result_dict)
    # Розгортання стовпця 'продажі'
    df_exploded = df.explode('продажі')
    # Розпакування словників у стовпці 'продажі' у окремі стовпці
    df_expanded = df_exploded.join(pd.json_normalize(df_exploded['продажі'])).drop(columns='продажі')
    # Сортування по кількість
    df_sorted = df_expanded.sort_values(by='кількість', ascending=False)

    #print(df_sorted)

    # Вказівка метаданих
    column_names = df_sorted.columns
    #print(column_names)

    month = df_sorted.iloc[0,0]
    year = df_sorted.iloc[0,1]

    with open("sorted_sales2.csv", 'w', newline='', encoding='utf-8') as csvfile:
        writer = csv.writer(csvfile)
        
        # Записуємо метадані
        writer.writerow([column_names[0], month])
        writer.writerow([column_names[1], year])
        
        # Записуємо порожній рядок
        writer.writerow([])
        
        # Записуємо заголовки для основних даних
        writer.writerow(df_sorted.columns[2:])
        
        # Записуємо основні дані
        for index, row in df_sorted.iterrows():
            writer.writerow(row[2:])

    # Запис метаданих в CSV файл

    
  

task_3_createCSV()


```

```python
#3 Створіть функцію, яка використовує бібліотеку pandas для створення DataFrame з CSV файлу,
#  виконує сортування даних за одним з стовпців і зберігає результат у новий CSV файл.
import pandas as pd
import csv

def task_3_createCSV(filename):

    sales_data = {
    'місяць': 'серпень',
    'рік': 2024,
    'продажі': [
        {'дата': '2024-08-01', 'продукт': 'Продукт A', 'кількість': 100, 'вартість': 5000},
        {'дата': '2024-08-02', 'продукт': 'Продукт B', 'кількість': 150, 'вартість': 7500},
        {'дата': '2024-08-03', 'продукт': 'Продукт C', 'кількість': 200, 'вартість': 10000},
        {'дата': '2024-08-04', 'продукт': 'Продукт D', 'кількість': 120, 'вартість': 6000},
        {'дата': '2024-08-05', 'продукт': 'Продукт E', 'кількість': 180, 'вартість': 9000},
        {'дата': '2024-08-06', 'продукт': 'Продукт F', 'кількість': 160, 'вартість': 8000},
        {'дата': '2024-08-07', 'продукт': 'Продукт G', 'кількість': 130, 'вартість': 6500},
        {'дата': '2024-08-08', 'продукт': 'Продукт H', 'кількість': 140, 'вартість': 7000},
        {'дата': '2024-08-09', 'продукт': 'Продукт I', 'кількість': 110, 'вартість': 5500},
        {'дата': '2024-08-10', 'продукт': 'Продукт J', 'кількість': 190, 'вартість': 9500}
    ],}


    with open(filename, 'w', newline='', encoding='utf-8') as csvfile:
        writer = csv.writer(csvfile)
        
        writer.writerow(['місяць', sales_data['місяць']])
        writer.writerow(['рік', sales_data['рік']])
        writer.writerow([])

        # заголовки таблиці
        writer.writerow(['дата', 'продукт', 'кількість', 'вартість'])

        # дані продажів
        for sale in sales_data['продажі']:
            writer.writerow([sale['дата'], sale['продукт'], sale['кількість'], sale['вартість']])


def task_3():
    filename = 'sales_data.csv'
    task_3_createCSV(filename)

    result_dict = dict()
    
    with open(filename, newline='', encoding='utf-8') as csvfile:
        reader = csv.reader(csvfile)

        for row in reader:
            if row == []:
                subdict_keys = next(reader)
                break
            result_dict[row[0]] = row[1]

    df_sales = pd.read_csv(filename, skiprows=4, header=None, encoding='utf-8')
    df_sales.columns = subdict_keys
    df_sorted = df_sales.sort_values(by='кількість', ascending=False)

    sorted_filename = 'sorted_sales_data.csv'

    with open(sorted_filename, 'w', newline='', encoding='utf-8') as csvfile:
        writer = csv.writer(csvfile)
        writer.writerow(['місяць', result_dict['місяць']])
        writer.writerow(['рік', result_dict['рік']])
        writer.writerow([]) 
        writer.writerow(df_sorted.columns)
        for index, row in df_sorted.iterrows():
            writer.writerow(row)

    print(df_sorted)


task_3()


```

```python
#9 Створіть функцію, що зчитує дані з CSV файлу, обробляє їх (наприклад, видаляє рядки з відсутніми значеннями), а потім зберігає результат у Parquet файл.
import csv
import numpy as np
import random

def create_csv():

    num_rows = 15
    num_columns = 5
    num_nans = 7

    data = np.random.randn(num_rows, num_columns)
    column_names = [f'Column_#{i+1}' for i in range(num_columns)]
    df = pd.DataFrame(data, columns=column_names)

    indices = [(r, c) for r in range(num_rows) for c in range(num_columns)]
    nan_positions = random.sample(indices, min(num_nans, len(indices)))

    for r, c in nan_positions:
        df.iat[r, c] = np.nan
    
    df.to_csv('task9.csv', index=False)
    



def task_10():
    create_csv()
    df = pd.read_csv('task9.csv')
    df_dropped = df.dropna()
    df_dropped.to_parquet('task_9.parquet', index=False)
    #print(df_dropped)

    df_read = pd.read_parquet('task_9.parquet')
    print(df_read)

    # return None

task_10()
```
